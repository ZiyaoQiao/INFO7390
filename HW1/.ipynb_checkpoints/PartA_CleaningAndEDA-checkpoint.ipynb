# Cleaning and EDA

## Data Cleaning

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats
import seaborn as sns
```

```python
file = pd.read_csv('Video_Games_Sales_as_at_22_Dec_2016.csv')
file.head()
```

The data contains 16 colums including the year, type of game, publisher, developer, sales number, score and rating. From the data displayed we can see some of the data is missing and we need to remove or impute these data.

### Are there missing value?

```python
pd.notnull(file).describe()
```

There are 16719 number of games and we can see from the describe that a lot of lines of data is missing the Critic_Score, Critic_Count, User_Score, User_Count, Developer and Rating, but the main part of the data(Sales) is complete. To make the analysis more concise and accurate, I would choose to drop the incomplete data.Below is the cleaning process.

```python
oldFile = file
file.duplicated()
file = file.drop_duplicates()
file = file.dropna()
pd.notnull(file).describe()
```

After the cleaning, there are 6825 sets of data left, which is around 40%. Though we lost a lot data, but it's enough and sufficient for the follwing analysis.

### Are there inappropraite values?

```python
check_total = file['NA_Sales'] + file['EU_Sales'] + file['JP_Sales'] + file['Other_Sales']
abs(check_total - file['Global_Sales']) <= 0.01
```

We can see from this check that some of the data is incorrect, the sales from each area is not the Global_sales provided(considering the error, +- 0.01). So we will correct the Global_sales with the sum of each area sales.

```python
file['Global_Sales'] = check_total
```

### How is the data distributed?

```python
%matplotlib inline
plt.figure(figsize=(18,6))
sns.distplot(file['Year_of_Release'])
```

```python
plt.figure(figsize=(18,6))
sns.stripplot(x="Year_of_Release", y="Global_Sales", data=file)
```

Most of the game sold is released during 2004 to 2015, and there is a strange data in 2006 which have a global_sales of over 80 which others are under 40. Regardless of this data which will discuss about later. The overall sales is quite consistent and the majority of game sold is in recent 10years.

### What are the summary statistics?

```python
file.describe()
```

The data contains game released between 1985 and 2016, The average of the Sales is 0.78, the average of the Critic score is 70 which is a reasonable.

```python
file.info()
```

There are 6825 sets of data is total and after the cleaning there is no Null value. Some of the data type are float64 and some of them are object.

### Are there anomalies/outliers?

```python
file[file.Global_Sales > 80]
```

From the above, we can see that the Wii sports sells more than 80 while others is less than 40. This is a significant outliers. According to the information about the Wii Sports, this game is sold with the console. Though this data is correct, but this uncommon game data will infect the analysis later, so we choose to ignore this data and use the remaining data for further study.

```python
file = file[file.Global_Sales < 80]
```

```python
file.sort_values(by = ['Global_Sales'])
file.head()
```

```python
file.to_csv('final.csv')
```

## Plots and Analytics

```python
plt.figure(figsize=(18,6))
sns.stripplot(x="Platform", y="Global_Sales", data=file)
```

This is a graph to show the Global Sales of each game in each platform, we can see most of the game with a high sales record is in Wii, DS, Xbox360, PS3 and PS2, the sales in other platform is low due to the low capacity of these consoles.

```python
data_Year_Sales = file.groupby(by=['Year_of_Release'],as_index = False)['Global_Sales'].sum()
data_Platform_Sales = file.groupby(by=['Platform'],as_index = False)['Global_Sales'].sum()
plt.figure(figsize=(18,5))
sns.barplot(x="Year_of_Release", y="Global_Sales", data=data_Year_Sales)
```

```python
plt.figure(figsize=(18,6))
sns.barplot(x="Platform", y="Global_Sales", data = data_Platform_Sales)
```

This is the overall Global Sales in each platforms, we can see PS2 PS3 Xbox360 Wii and DS have the highest overall Global Sales. It seems that though PS2 have a very long history and is supposed to be stopped releasing new games, it still sales a lot of games due to it's large capacities. The new generation PS4 and XboxOne sales lower than expected.

```python
plt.figure(figsize=(18,6))
data_Genre = file.groupby([file['Year_of_Release'],file['Genre']],as_index = False)['Global_Sales'].sum()
sns.barplot(x='Genre',y='Global_Sales', data=data_Genre)
```

```python
plt.figure(figsize=(18,6))
data_Genre = file.groupby([file['Year_of_Release'],file['Genre']],as_index = False)['Global_Sales'].mean()
sns.barplot(x='Genre',y='Global_Sales', data=data_Genre)
```

The above two graphs show the preference of people buying games. Action, Shooter and Sports are the top 3 popular genre. According to the second graph which is the mean of each Genre, the averge is about to be the same, no significantly differ between these three genres.

```python
plt.figure(figsize=(18,6))
data_Genre = file.groupby([file['Year_of_Release'],file['Rating']],as_index = False)['Global_Sales'].sum()
sns.countplot(x='Rating', data=data_Genre)
```

```python
plt.figure(figsize=(18,5))
data_Genre = file.groupby([file['Year_of_Release'],file['Rating']],as_index = False)['Global_Sales'].mean()
sns.barplot(x='Rating',y='Global_Sales', data=data_Genre)
```

```python
labels = 'North America Sales','Europe Sales','Japan Sales','Total'
sizes = [file['NA_Sales'].mean()/file['Global_Sales'].mean()*100,file['EU_Sales'].mean()/file['Global_Sales'].mean()*100,file['JP_Sales'].mean()/file['Global_Sales'].mean()*100,file['Other_Sales'].mean()/file['Global_Sales'].mean()*100]
explode = (0,0,0,0)
plt.pie(sizes,explode=explode,labels=labels,autopct='%1.1f%%',shadow=False,startangle=90)
plt.axis('equal')
plt.show()
```

This chart shows how the sales is distributed. North America sales more than half of the games while Europe and Japan sales about 40%. The other regions sales only 10% of total sales.
Most of the game sold is T,E10+, M and E, because most of the game sold is in North America, and these four Ratings are america standard.

```python
sns.pairplot(file, vars=['Year_of_Release', 'Critic_Score','Global_Sales'])
```

This is the distribution of Global_Sales, Critic_Score and Year_of Release, we can see the distribution between these three variables.The bottom-middle graph is more interesting, it seems that the higher Critic Score might tends to have higher Global sales.

```python
data_Year_perSales = file[['Year_of_Release','Global_Sales','Critic_Score']]
data_Year_2016 = data_Year_perSales[data_Year_perSales['Year_of_Release'] == 2016]
sns.pairplot(data_Year_2016, vars=[ 'Critic_Score','Global_Sales'])
```

```python
data_Year_2015 = data_Year_perSales[data_Year_perSales['Year_of_Release'] == 2015]
sns.pairplot(data_Year_2015, vars=[ 'Critic_Score','Global_Sales'])
```

The above two plots are the relationship betwwen Critic Score and Global Sales in year 2016 and 2015. This plot shows clear evidence that the higher critic score have a higher potential to sale more.

```python
plt.figure(figsize=(18,6))
tmpFile = file[file['Year_of_Release'] > 2010]
sns.violinplot(x='Year_of_Release',y='Critic_Score',data=tmpFile)
```

The critic score of each year is relatively stable, this means the average level of games doesn't change a lot in these years.

```python
plt.figure(figsize=(18,6))
sns.violinplot(x='Platform',y='Critic_Score',data=file)
```

This shows the score over different platform. The data of DC platform is a little strange, this may because there is less game in this platform and influence the distribution. Other platforms have relatively similar result, basicaly between 60 to 90.

```python
plt.figure(figsize=(18,6))
file['User_Score'] = file['User_Score'].astype('float64')
file['Diff_Between'] = abs(file['Critic_Score'] - file['User_Score']*10)
data_Year_Diff = file.groupby(by=['Year_of_Release'],as_index = False)['Diff_Between'].mean()
sns.regplot(x="Critic_Score", y="Diff_Between", data=file)
```

This is a plot shows the diviation between user critic and media critic score. It seems the higher the score is, the more likely the two score become closer. This suggest a phenomenon that to choose a game, if its score is high, you can arbitrary follow the media while if its score is low, the game is mainly focus on small parts of players and parts of the players may fall in love with it while others may hate it.

Are any of the columns correlated?

```python
cmap = sns.diverging_palette(0, 255, sep=1, n=256, as_cmap=True)

correlations = file[['Global_Sales', 'Critic_Score','User_Score', 'Year_of_Release']].corr()
print (correlations)
sns.heatmap(correlations, cmap=cmap)
```

This is the correlation between these four facts. It seems the relationship is quite weak and only some relationship between UserScore and Critic Score. The above will have a more detail analysis about this.

```python
file['Genre'] = file['Genre'].astype('category')
file_Action = file[file['Genre']=="Action"]
file_Shooter = file[file['Genre']=="Shooter"]
file_Role_Playing = file[file['Genre']=="Role-Playing"]
file_Sports = file[file['Genre']=="Sports"]
file_Racing = file[file['Genre']=="Racing"]

cmap = sns.diverging_palette(0, 255, sep=1, n=256, as_cmap=True)
correlations = file_Action[['Global_Sales', 'Critic_Score','User_Score', 'Year_of_Release']].corr()
print (correlations)
sns.heatmap(correlations, cmap=cmap)
```

```python
correlations = file_Shooter[['Global_Sales', 'Critic_Score','User_Score', 'Year_of_Release']].corr()
print (correlations)
sns.heatmap(correlations, cmap=cmap)
```

```python
correlations = file_Role_Playing[['Global_Sales', 'Critic_Score','User_Score', 'Year_of_Release']].corr()
print (correlations)
sns.heatmap(correlations, cmap=cmap)
```

```python
correlations = file_Sports[['Global_Sales', 'Critic_Score','User_Score', 'Year_of_Release']].corr()
print (correlations)
sns.heatmap(correlations, cmap=cmap)
```

```python
correlations = file_Racing[['Global_Sales', 'Critic_Score','User_Score', 'Year_of_Release']].corr()
print (correlations)
sns.heatmap(correlations, cmap=cmap)
```

The above plots reveals the correlation between Global Sales, Critic Score, User Score and Year of Release over different Genre of Game. From these plots, we can see the Role Playing Game(RPG) has a higher correlation between score and Global Sales. While the correlation is smaller for other type of game. Action and Shooter game is a slight behind RPG game while Sports and Racing have much smaller correlation. It seems those who buy sports or Racing game are not concerning about the score. From my own experience corresponding to this result, Sports or Racing game players are more likely to buy games depending on the game ip.

From the above, we can see that Global Sales is correlated with Critic Score, and the Critic Score is also correlated with User Score. Also the Year of Release have some influence on these columns.
