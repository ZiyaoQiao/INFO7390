### Remove or impute any bad data. 

```python
fp = fp.dropna(how='any')
len(fp)
```

There is **41757 rows is complete and available for analyze**, it is really a good new which means most of data is complete and conseutive, really ideal for analyze.

```python
def label_season (row):
    if row['month'] == 3 or row['month'] == 4 or row['month'] == 5:
      return 'Spring'
    if row['month'] == 6 or row['month'] == 7 or row['month'] == 8:
      return 'Summer'
    if row['month'] == 9 or row['month'] == 10 or row['month'] == 11:
      return 'Autumn'
    if row['month'] == 12 or row['month'] == 1 or row['month'] == 2:
      return 'Winter'
fp['season'] = fp.apply (lambda row: label_season (row),axis=1)
fp = fp.drop('No', axis=1)
fp.to_csv('out.csv')
```

Since the original data have month data and is time series, I could analyze the dataset base on different season. I just <span style="color:blue">**add 'season' as a new column base on different months.**</span> Besides, since <span style="color:red">the column 'No' is meanless,</span> I prefer drop it and instead with datetime in the future. <span style="color:green">** Finally, I output a new csv based on cleaned and tackled data, if specialists to improve the dataset or vaildate it, they could just use the csv without additional operations.**</span>

### Additional transform: form daily dataset according to original hourly dataset.

```python
days = (file.index[0] + timedelta(i) for i in range((file.index[-1] - file.index[0]).days + 1)) # collect all day data
columns = ['pm2.5', 'DEWP', 'PRES', 'TEMP', 'Iws', 'Is', 'Ir']

def dailygather(dp, column, pre):
    indexies = ['mean', 'min', 'max']
    file_desc = file.loc[dp: dp + timedelta(hours=23)].describe()
    s = file_desc[column][indexies]
    s.index = s.index.map(lambda x: '-'.join((column, x)))
    return s

pre = [0]
file_daily = pd.DataFrame({dt: pd.concat([dailygather(dt, column, pre) for column in columns]) for dt in list(days)}).T
```

```python
file_daily_mean = file_daily[['{}-mean'.format(column) for column in columns]]
file_daily_mean.describe()
```

Complete the daily dataset should be helpful because it would make the dataset much more stable, decrease the vibrate caused by some exterme data.

```python
correlations_daily = file_daily_mean[['pm2.5-mean', 'DEWP-mean', 'PRES-mean', 'TEMP-mean', 'Iws-mean', 'Is-mean', 'Ir-mean']].corr()
correlations_daily
```

```python
cmap = sns.diverging_palette(0, 255, sep=16, n=1024, as_cmap=True)
sns.heatmap(correlations_daily, cmap=cmap)
```

Really similar with hourly data but also have some tiny difference, such as the relation between rainy hour and dewp, which become more apparent by utilizing daily perspective.
